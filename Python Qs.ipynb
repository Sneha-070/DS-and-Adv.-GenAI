### **Basic Python Programming**

```python
#printing a string 

str = "Hello, World!"
print(str)
```

```python
#python if-else

n = int(input().strip())

if n % 2 != 0:
    print("Weird")
else:
    if 2 <= n <= 5:
        print("Not Weird")
    elif 6 <= n <= 20:
        print("Weird")
    else:
        print("Not Weird")
```

```python
#arithmetic operators

a = int(input())
b = int(input())

print(a+b)
print(a-b)
print(a*b)
```

```python
#python : division

a = int(input())
b = int(input())

print(a//b)
print(a/b)
```

```python
#loops

n = int(input())

for i in range (n):
    print(i*i)
```

```python
#function

def is_leap(year):
    if year % 400 == 0:
        return True
    if year % 100 == 0:
        return False
    if year % 4 == 0:
        return True
    return False


year = int(input())
print(is_leap(year))
```

```python
#print function

n = int(input())

for i in range(1, n+1):
    print(i, end="")
```
